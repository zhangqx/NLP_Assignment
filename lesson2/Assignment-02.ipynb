````python
import matplotlib
%matplotlib inline
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import re


url = 'https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
loc = url
# 加上useragent，否则百度会返回重定向
kv = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64; rv:44.0) Gecko/20100101 Firefox/44.0"}
response = requests.get(loc, headers=kv, allow_redirects=False

response.encoding = 'utf-8'
html = response.text
# 得到整个页面的结构
soup = BeautifulSoup(html, "html.parser")
soup
````

````python
# 获取所有地铁线路
tables = soup.find_all(name='table')
links = []    # 用于保存所有地铁线路
for table in tables:
    ths = table.find_all('th')
    for th in ths:
        if th.string == '线路号':
            links = table.find_all('a')
            print(links)
            break
    
````

````python
# 根据每条线路的站点生成站点地图
def make_station_map(stations):
    for i, item in enumerate(stations):
        if i + 1 >= len(stations): 
            station_map[item].append(stations[i-1])
            break
        if i == 0:
            station_map[item].append(stations[i+1])
        else:
            station_map[item].append(stations[i-1])
            station_map[item].append(stations[i+1])
     
````

````python
# 获取当前url对应的地铁线路所有的站点
def get_stations(url):
    print('url: ', url)
    res = requests.get(url, headers=kv, allow_redirects=False)
    res.encoding = 'utf-8'
    linehtml = res.text
    # 得到当前线路的页面结构
    linesoup = BeautifulSoup(linehtml, 'html.parser')
    linetables = linesoup.find_all('table')
    stationlist = []
    for table in linetables:
        if table.caption :
            print(table.caption)
            text = table.caption.string
            print(text)
            pattern = re.compile('车站列表')
            if text and pattern.search(text):
                trs = table.find_all('tr')
                for tr in trs:
                    if tr.a:
                        stationlist.append(tr.a.string)
                print(stationlist)
    return stationlist
    
````

````python
get_stations('https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%8110%E5%8F%B7%E7%BA%BF')
````

````python
# 遍历地铁线路，生成站点地图
url_prefix = 'https://baike.baidu.com'
pattern = re.compile(r'href="(/item/*)"')
station_map = defaultdict(list)
for link in links:
    if link.has_attr('href'):
        stationlist = get_stations(url_prefix + link['href'])
        make_station_map(stationlist)
print(station_map)
````

````python
station_map['望京站']
````

````python
def search(mstation_map, start, end):
    pathes = [[start]]
    seen = []
    while pathes:
        path = pathes.pop(0)
        froniter = path[-1]
        
        if froniter in seen: continue
            
        nextes = mstation_map[froniter]
        
        for nextone in nextes:
            if nextone in path: continue
            
            new_path = path+ [nextone]
            pathes.append(new_path)
            
            if nextone == end: return new_path
        seen.append(froniter)
        
        
````

````python
search(station_map, "望京站", "北京南站")
````
